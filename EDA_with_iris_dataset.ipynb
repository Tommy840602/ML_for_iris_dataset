```python
import pandas as pd
import warnings
import seaborn as sns
import matplotlib.pyplot as plt

iris = pd.read_csv("Iris.csv")
```

```python
iris.head()
```

```python
iris["Species"].value_counts()
```

```python
#Data Visualizations
iris.plot(kind="scatter", x="SepalLengthCm", y="SepalWidthCm")
```

```python
sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=iris)
```

```python
sns.FacetGrid(iris,hue="Species").map(plt.scatter, "SepalLengthCm", "SepalWidthCm").add_legend()
```

```python
ax = sns.boxplot(x="Species", y="PetalLengthCm", data=iris)
ax = sns.stripplot(x="Species", y="PetalLengthCm", data=iris, jitter=True, edgecolor="gray")
```

```python
sns.violinplot(x="Species", y="PetalLengthCm", data=iris)
```

```python
sns.FacetGrid(iris, hue="Species").map(sns.kdeplot,"PetalLengthCm").add_legend()
```

```python
sns.pairplot(iris.drop("Id", axis=1), hue="Species",diag_kind="kde")
```

```python
iris.drop("Id", axis=1).boxplot(by="Species", figsize=(12, 6))
```

```python
#Andrews curves
pd.plotting.andrews_curves(iris.drop("Id", axis=1), "Species",)
```

```python
#Parallel coordinates
pd.plotting.parallel_coordinates(iris.drop("Id", axis=1), "Species")
```

```python
#RadViz
pd.plotting.radviz(iris.drop("Id", axis=1), "Species")
```
